```python
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)

# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
df
```

```python
data = {
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30],
    'Gender': ['F', 'M']
}

df = pd.DataFrame(data)
df.groupby('Gender')['Age'].mean()
df
```
